```python
import pandas as pd
import numpy as np
import random
```

```python
recon = pd.read_csv("download/processed/recon3M.csv")
labels = pd.read_csv("download/processed/labels3M.csv")
labels = labels[["pT","local_y"]]
```

```python
recon.shape,labels.shape
```

```python
labels["pT"] = labels.apply(lambda x: 1.0 if x["pT"] >= 2.0 else 0.0, axis = 1)
```

```python
major_idxs, minor_idxs = labels[labels["pT"] == 0.0].index, labels[labels["pT"] == 1.0].index
minor_idxs.shape, major_idxs.shape
```

```python
signal_idxs = random.choices(minor_idxs, k=100000)
background_idxs = np.random.choice(major_idxs, size = 100000, replace = False)
len(signal_idxs), len(background_idxs)
```

```python
signal_labels = labels.loc[signal_idxs,:]
background_labels = labels.loc[background_idxs,:]
signal_recon = recon.loc[signal_idxs,:]
background_recon = recon.loc[background_idxs,:]
```

```python
combined_recon = pd.concat([signal_recon, background_recon],axis = 0)
combined_recon.head()
```

```python
combined_labels = pd.concat([signal_labels, background_labels],axis = 0)
combined_labels.head()
```

```python
df1 = pd.concat([combined_recon,combined_labels],axis = 1)
df1.shape
```

```python
df1.sort_index(ascending=True, inplace=True)
```

```python
df1
```

```python
df1.to_csv(path_or_buf="download/processed/resampled200K.csv", index= False)
```
