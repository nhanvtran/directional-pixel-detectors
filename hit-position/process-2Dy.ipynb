```python
import numpy as np
import pandas as pd
import os
import math
```

```python
# take first 8 time slices
cols = list(range(0,13*8)) 
```

```python
eosdir_pos = '/eos/uscms/store/user/jennetd/smart-pixels/dataset678/unflipped-positive/'
eosdir_neg = '/eos/uscms/store/user/jennetd/smart-pixels/dataset678/unflipped-negative/'

df = pd.DataFrame()

i=0
print('Processing + charge')
for f in os.listdir(eosdir_pos):

    if 'recon' in f:
        continue
    
    i = i + 1
    if i > 150:
        break
            
    label = pd.read_csv(eosdir_pos+f,usecols=['y-midplane'])
                        
    xdf = pd.read_csv(eosdir_pos+'recon2Dy_'+f.split('_')[1],usecols=cols)
    
    df = pd.concat([df,pd.concat([label,xdf],axis=1)])
    
i=0
print('Processing - charge')
for f in os.listdir(eosdir_neg):
    
    if 'recon' in f:
        continue
        
    i = i + 1
    if i > 150:
        break

    label = pd.read_csv(eosdir_neg+f,usecols=['y-midplane'])
                        
    xdf = pd.read_csv(eosdir_neg+'recon2Dy_'+f.split('_')[1],usecols=cols)
    
    df = pd.concat([df,pd.concat([label,xdf],axis=1)])
```

```python
df.head()
```

```python
shuffled_df = df.sample(frac=1).reset_index(drop=True)
shuffled_df.head()
```

```python
split_n = math.floor(0.8*len(shuffled_df))
```

```python
train_df = shuffled_df[:split_n]
test_df = shuffled_df[split_n:]

print(len(train_df),len(test_df))
```

```python
train_df.to_csv('train_2Dy.csv',index=False)
test_df.to_csv('test_2Dy.csv',index=False)
```
