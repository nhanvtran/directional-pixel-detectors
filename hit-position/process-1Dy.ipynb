```python
import numpy as np
import pandas as pd
import os
import math
```

```python
eosdir_pos = '/eos/uscms/store/user/jennetd/smart-pixels/dataset678/unflipped-positive/'
eosdir_neg = '/eos/uscms/store/user/jennetd/smart-pixels/dataset678/unflipped-negative/'

df = pd.DataFrame()

print('Processing + charge')
for f in os.listdir(eosdir_pos):
    if 'recon' in f:
        continue
        
    label = pd.read_csv(eosdir_pos+f,usecols=['y-midplane','cotBeta'])
                        
    ydf = pd.read_csv(eosdir_pos+'recon1Dy_'+f.split('_')[1])
    
    df = pd.concat([df,pd.concat([label,ydf],axis=1)])
    
print('Processing - charge')
for f in os.listdir(eosdir_neg):
    if 'recon' in f:
        continue

    label = pd.read_csv(eosdir_neg+f,usecols=['y-midplane','cotBeta'])
                        
    ydf = pd.read_csv(eosdir_neg+'recon1Dy_'+f.split('_')[1])
    
    df = pd.concat([df,pd.concat([label,ydf],axis=1)])
```

```python
df.head()
```

```python
shuffled_df = df.sample(frac=1).reset_index(drop=True)
shuffled_df.head()
```

```python
split_n = math.floor(0.8*len(shuffled_df))
```

```python
train_df = shuffled_df[:split_n]
test_df = shuffled_df[split_n:]

print(len(train_df),len(test_df))
```

```python
train_df.to_csv('train_1Dy.csv',index=False)
test_df.to_csv('test_1Dy.csv',index=False)
```
